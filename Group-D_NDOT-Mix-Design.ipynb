```python
# Defining constants and adding relevant information
project_name = "Project #2: Automation of Concrete Mix Design for NDOT"
client = "Nebraska Department of Transportation (NDOT)"


cubic_yard_ft3 = 27      # Total volume in cubic feet for 1 cubic yard
unit_ww = 62.4           # Unit weight of water (lb/ft^3)

# Specific gravities

sg_cement_J = 3.15       # Specific gravity of cement (J)
sg_flyash_K = 2.30       # Specific gravity of fly ash (K)
sg_silicafume_L = 2.20   # Specific gravity of silica fume (L)
sg_SCM_M = 2.60          # Specific gravity of other SCM (M)
```

```python
# Function to calculate water weight (Q)
def water_Q(cement_w_A, fly_ash_w_B, silica_fume_w_C, other_SCM_w_D, WC_ratio_E):
    
    cement_total = cement_w_A + fly_ash_w_B + silica_fume_w_C + other_SCM_w_D  
    # Sum of all material weights
    
    water_weight_Q = cement_total * WC_ratio_E  
    # Multiply by water-cement ratio
    
    return water_weight_Q  
    # Returns water weight in pounds

#Q = water_Q(cement_w_A, fly_ash_w_B, silica_fume_w_C, other_SCM_w_D, WC_ratio_E)
# Calculates total water weight

#print(f"Water Weight (Q): {Q:.2f} lb")
```

```python
# Function to calculate cement volume (R)
def volume_R(cement_w_A, sg_cement_J):
    
    volume_R = cement_w_A / (sg_cement_J * unit_ww)  
    # Divides cement weight by (specific gravity × unit weight of water)
    
    return volume_R  
    # Returns cement volume in cubic feet
    
#R = volume_R(cement_w_A, sg_cement_J)
# Calculates cement volume

#print(f"Cement Volume (R): {R:.3f} ft^3")

```

```python
# Function to calculate fly ash volume (S)
def volume_S(fly_ash_w_B, sg_flyash_K):
    
    volume_S = fly_ash_w_B / (sg_flyash_K * unit_ww)  
    # Divides fly ash weight by (specific gravity × unit weight of water)
    
    return volume_S  
    # Returns fly ash volume in cubic feet
#S = volume_S(fly_ash_w_B, sg_flyash_K)
# Calculates fly ash volume

#print(f"Fly Ash Volume (S): {S:.3f} ft^3")
```

```python
# Function to calculate silica fume volume
def volume_T(silica_C, sg_silicafume_L):

    volume_T = silica_C / (sg_silicafume_L * unit_ww)
    #divides the silica fume weight by (specific gravity * unit weight of water)

    return volume_T
    # Returns the value of the silica fume volume in cubic feet
```

```python
#Function to calculate the other SCM volume
def volume_U(other_scm_D, sg_SCM_M):

    volume_U = other_scm_D / (sg_SCM_M * unit_ww)
    #divides the scm weight by (specific gravity * unit weight of water)
    
    return volume_U
    #Returns the value of the other SCM volume in cubic feet
```

```python
#Function to calculate the air volume
def volume_V(target_air_content_F):
    
    volume_air = (target_air_content_F/100) * cubic_yard_ft3
    #divides the target air percent by 100 and then that value is multiplied
    
    return volume_air
    #Returns the volume of the air in cubic feet
```

```python
# Define a function to calculate water volume (W) as a function of water weight (Q)
def volume_W(weight_water_Q):
    
    water_volume = weight_water_Q / unit_ww
    #equation to find the volume of water 
    
    return water_volume
    #Returns the volume of the water in cubic feet
```

```python
# Define a function to calculate total aggregate volume (X)
def volume_X(volume_R,volume_S,volume_T,volume_U,volume_V,volume_W):
    
    volume_X = (27 - volume_R - volume_S - volume_T - volume_U - volume_V - volume_W)
    #equation that finds the total aggregate volume
    
    return volume_X
    #returns the total aggregate volume in cubic feet
```

```python
# Define a function to calculate fine aggregate weight (Y)
def weight_Y(target_G,sg_fine_aggregate_N,aggregate_volume_X):
    
    fine_aggregate_weight_Y = 62.4 * (target_G/100) * sg_fine_aggregate_N * aggregate_volume_X
    #equation that finds just the weight of the fine aggregate included in the mix
    
    return fine_aggregate_weight_Y
    #returns the fine aggregate weight in pounds
```

```python
# Function for coarse aggregate weight (Z)
def weight_Z(target_H,sg_coarse_aggregate_O,aggregate_volume_X):
    
    coarse_aggregate_weight_Z = 62.4 * (target_H/100) * sg_coarse_aggregate_O * aggregate_volume_X
    #equation that finds just the weight of the coarse aggregate in the mix
    
    return coarse_aggregate_weight_Z
    #returns the value of the weight of coarse aggregate in pounds
```

```python
# Function for other aggregate weight (AA)
def weight_AA(target_I,sg_other_aggregate_P,aggregate_volume_X):
    
    other_aggregate_weight_AA = 62.4 * (target_I/100) * sg_other_aggregate_P * aggregate_volume_X
    #equation that finds just the weight of other aggregate in the mix
    
    return other_aggregate_weight_AA
    #returns the value of the other aggregate weight in pounds
```

```python
#NDOT Concrete Mix Design - Full Calculation

#metadata
project_number = int(input("Enter project number: "))
class_of_concrete = input("Enter class of concrete (e.g., 47B, 47BR: ")

#user inputs for the cementitious materials (all in lb per cubic yard)

cement_A = float(input("Enter cement weight A (lb per cubic yard): "))
fly_ash_B = float(input("Enter fly ash weight B (lb per cubic yard): "))
silica_fume_C = float(input("Enter silica fume weight C (lb per cubic yard): "))
other_scm_D = float(input("Enter other SCM weight D (lb per cubic yard): "))

#Water cement ratio and air content targets inputs for the concrete

water_cement_ratio_E = float(input("Enter target water-cement ratio E: "))
air_content_F = float(input("Enter target air content F (%): "))

# Aggregate proportions (%)

percent_fine_G = float(input("Enter percent fine aggregate G (%): "))
percent_coarse_H = float(input("Enter percent coarse aggregate H (%): "))
percent_other_I = float(input("Enter percent other aggregate I (%): "))

#User inputs for the specific gravities of the materials

sg_cement_J = float(input("Enter specific gravity of cement J: "))
sg_fly_ash_K = float(input("Enter specific gravity of fly ash K: "))
sg_silica_fume_L = float(input("Enter specific gravity of silica fume L: "))
sg_other_scm_M = float(input("Enter specific gravity of other SCM M: "))

sg_fine_N = float(input("Enter specific gravity of fine aggregate N: "))
sg_coarse_O = float(input("Enter specific gravity of coarse aggregate O: "))
sg_other_P = float(input("Enter specific gravity of other aggregate P: "))

#Calculations based off of the user inputs - calling functions previously defined

Q = water_Q(cement_A, fly_ash_B, silica_fume_C, other_scm_D, water_cement_ratio_E)

R = volume_R(cement_A, sg_cement_J)
S = volume_S(fly_ash_B, sg_fly_ash_K)
T = volume_T(silica_fume_C, sg_silica_fume_L)
U = volume_U(other_scm_D, sg_other_scm_M)

V = volume_V(air_content_F)
W = volume_W(Q)

X = volume_X(R,S,T,U,V,W)

Y = weight_Y(percent_fine_G, sg_fine_N, X)
Z = weight_Z(percent_coarse_H, sg_coarse_O, X)
AA = weight_AA(percent_other_I, sg_other_P, X)

# Output: Final Weight Chart
print("\n---------------------------------------------")
print(" NDOT Concrete Mix Design - Weight Summary")
print("         (1 Cubic Yard of Concrete)")
print("---------------------------------------------")
print(f"Project Number:       {project_number}")
print(f"Class of Concrete:    {class_of_concrete}")
print("---------------------------------------------")
print(f"Cement (A):            {cement_A:8.1f} lb")
print(f"Fly Ash (B):           {fly_ash_B:8.1f} lb")
print(f"Silica Fume (C):       {silica_fume_C:8.1f} lb")
print(f"Other SCM (D):         {other_scm_D:8.1f} lb")
print("---------------------------------------------")
print(f"Fine Aggregate (Y):    {Y:8.0f} lb")
print(f"Coarse Aggregate (Z):  {Z:8.0f} lb")
print(f"Other Aggregate (AA):  {AA:8.0f} lb")
print("---------------------------------------------")
print(f"Water (Q):             {Q:8.0f} lb")
print("---------------------------------------------")
print("End of Mix Design Summary")
```
